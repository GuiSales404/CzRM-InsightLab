```python
df = train_df_to_adjust.drop([
                            'Unnamed: 0',
                            'A senhora ou sua família recebem algum benefício do governo do Estado do Ceará (poderá marcar mais de uma resposta)_x',
                            'Nos últimos 12 meses, alguém no seu domicílio fez algum curso de qualificação?',
                            'Qual a forma de iluminação utilizada no domicílio? (marque o mais utilizado)',
                            ], axis=1)
```

```python
df
```

```python
X = df.iloc[:,1:26]
y = df['Y']
```

```python
# grave:0 | leve:1 | moderada:2 | sem ia: 3
```

###Auto Balanced Test


```python
from imblearn.combine import SMOTETomek 
smt = SMOTETomek(random_state=42)

X_res, y_res = smt.fit_resample(X, y)
```

```python
analysis = pd.DataFrame(y_res, columns = ['ia level'])
```

```python
grave = 0
leve = 0 
moderada = 0 
sem_ia = 0

for x in analysis['ia level']:
    if x == 'grave':
        grave += 1
    if x == 'leve':
        leve += 1
    if x == 'moderada':
        moderada += 1
    if x == 'sem ia': 
        sem_ia += 1

print(f'Grave: {grave}\nLeve: {leve} \nModerada: {moderada} \nSem ia:  {sem_ia} ')
```

```python
df = X_res
df['ia level'] = y_res
df
```

```python
df.to_csv('/content/drive/MyDrive/Repasse Gui/df_auto_balanced_labeled.csv')
```

```python
# grave:0 | não grave:1 
```

```python
df['Y'] = df['Y'].replace([2, 3], 1)
```

```python
X = df.iloc[:, 9:26]
y = df['Y']
```

## Auto Balanced

```python
from imblearn.combine import SMOTETomek 
smt = SMOTETomek(random_state=42)

X_res, y_res = smt.fit_resample(X, y)
```

```python
grave = 0
n_grave = 0

for x in y_res:
    if x == 0:
        grave += 1
    if x == 1:
        n_grave += 1


print(f'Grave: {grave}\nNão Grave: {n_grave}')
```

```python
results = X_res
results['ia level'] = y_res
```

```python
results.to_csv('/content/drive/MyDrive/Repasse Gui/dataset_balanced/mean question/auto_balanced_2C')
```
